## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```python
# File location and type
file_location = "/FileStore/tables/Absenteeism_at_work2.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

#Data Preprocessing

```python
# Create a 70-30 train test split

train_data,test_data=df.randomSplit([0.7,0.3])
```

```python
# Import the required libraries

from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import GeneralizedLinearRegression
from pyspark.ml.feature import VectorAssembler,StringIndexer,StandardScaler
from pyspark.ml import Pipeline
```

```python
# Use StringIndexer to convert the categorical columns to hold numerical data

month_of_absence_indexer = StringIndexer(inputCol="Month of absence",outputCol='month_of_absence_index',handleInvalid='keep')
day_of_the_week_indexer = StringIndexer(inputCol="Day of the week",outputCol="day_of_the_week_index",handleInvalid='keep')
son_indexer = StringIndexer(inputCol="Son",outputCol='son_index',handleInvalid='keep')
pet_indexer = StringIndexer(inputCol="Pet",outputCol="pet_index",handleInvalid='keep')
Seasons_indexer = StringIndexer(inputCol="Seasons",outputCol='Seasons_index',handleInvalid='keep')
```

```python
assembler = VectorAssembler(inputCols=["Transportation expense","Distance from Residence to Work","Service time","Age","workload_avg ","Hit target","Disciplinary failure","Education","Social drinker","Body mass index"],
                           outputCol="unscaled_features")
```

```python
# Standard scaler is used to scale the data for the linear SVC to perform well on the training data

scaler = StandardScaler(inputCol="unscaled_features",outputCol="features")
```

```python
pipe = Pipeline(stages=[month_of_absence_indexer,day_of_the_week_indexer,Seasons_indexer,son_indexer,pet_indexer,assembler,scaler])
```

##Models

Linear Regression

```python
fitted_pipe=pipe.fit(train_data)
```

```python
train_data_lm=fitted_pipe.transform(train_data)
```

```python
# Create an object for the Linear Regression model

lr_model = LinearRegression(labelCol='Absenteeism time in hours')
```

```python
# Fit the model on the train data

fit_model = lr_model.fit(train_data_lm.select(['features','Absenteeism time in hours']))
```

```python
print("Coefficients: %s" % str(fit_model.coefficients))
print("Intercept: %s" % str(fit_model.intercept))
```

```python
#dir(fit_model.summary)
print(fit_model.summary.pValues)
```

```python
trainingSummary = fit_model.summary
print("numIterations: %d" % trainingSummary.totalIterations)
print("objectiveHistory: %s" % str(trainingSummary.objectiveHistory))
trainingSummary.residuals.show()
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```python
# Transform the test data using the model to predict the duration

test_data_lm=fitted_pipe.transform(test_data)
```

```python
# Store the results in a dataframe
2
 
3
results = fit_model.transform(test_data_lm)
```

```python
from pyspark.sql.functions import when
results=results.withColumn("prediction", when(results.prediction < 0, 0)
                           .otherwise(results.prediction))
```

```python
from pyspark.sql.functions import round, col

results.select('Absenteeism time in hours', round(col('prediction'))).display()
```

```python
test_results = fit_model.evaluate(test_data_lm)
```

```python
test_results.residuals.show()
```

```python
test_results.rootMeanSquaredError
```

```python
test_results.r2
```

Generalized Linear Regression

```python
fitted_pipe=pipe.fit(train_data)
```

```python
train_data_glm=fitted_pipe.transform(train_data)
```

```python
glr = GeneralizedLinearRegression(labelCol='Absenteeism time in hours',family="poisson", link="log", maxIter=10, regParam=0.3)
fit_model = glr.fit(train_data_glm.select(['features','Absenteeism time in hours']))
```

```python
print("Coefficients: " + str(fit_model.coefficients))
print("Intercept: " + str(fit_model.intercept))
```

```python
# Summarize the model over the training set and print out some metrics
summary = fit_model.summary
print("Coefficient Standard Errors: " + str(summary.coefficientStandardErrors))
print("T Values: " + str(summary.tValues))
print("P Values: " + str(summary.pValues))
print("Dispersion: " + str(summary.dispersion))
print("Null Deviance: " + str(summary.nullDeviance))
print("Residual Degree Of Freedom Null: " + str(summary.residualDegreeOfFreedomNull))
print("Deviance: " + str(summary.deviance))
print("Residual Degree Of Freedom: " + str(summary.residualDegreeOfFreedom))
print("AIC: " + str(summary.aic))
print("Deviance Residuals: ")
summary.residuals().show()
```

```python
# Transform the test data using the model to predict the duration

test_data_glm=fitted_pipe.transform(test_data)
```

```python
# Store the results in a dataframe

results = fit_model.transform(test_data_glm)
```

```python
from pyspark.sql.functions import round, col

results.select('Absenteeism time in hours', round(col('prediction'))).display()
```

```python
test_results = fit_model.evaluate(test_data_glm)
```
